```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
phi = np.array([0.7, 0.7])
theta = np.array([0.4, -0.4])
k = np.arange(1, 20)

rho = lambda p, t: (1 - p * t) * (p - t) / (1 - 2 * p * t + t**2)
for i in range(len(phi)):
    plt.plot(k, rho(phi[i], theta[i]) * phi[i]**k)

plt.legend(['phi = '  + str(phi[0]) + ', theta = ' + str(theta[0]), 
           'phi = '  + str(phi[1]) + ', theta = ' + str(theta[1])])
plt.title('Autocorrelation Function for ARMA(1, 1)')
plt.ylabel('Autocorrelation Function')
plt.xlabel('Lag, k')
plt.show()
```
